```python
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def load_dicom_volumes(folder_path):
    volumes = []

    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            slices = [pydicom.dcmread(os.path.join(subfolder_path, file)) for file in os.listdir(subfolder_path) if file.lower().endswith('.dcm')]
            slices.sort(key=lambda x: float(x.ImagePositionPatient[2]))

            volume = np.stack([s.pixel_array for s in slices], axis=-1)
            volumes.append(volume)

    return volumes

def visualize_slices(volume, z_slice):
    plt.imshow(volume[:, :, z_slice], cmap='gray', origin='lower', aspect='auto')
    plt.title(f'Slice at Z = {z_slice}')
    plt.show()

# 대상 디렉터리 지정
patient_id = '0090122'
ct_directory = os.path.join(r'F:\up_under_arm\upper\ct', patient_id)
pet_directory = os.path.join(r'F:\up_under_arm\upper\pet', patient_id)

# CT 및 PET 볼륨 로드
ct_volumes = load_dicom_volumes(ct_directory)
pet_volumes = load_dicom_volumes(pet_directory)

# 상호 작용적인 시각화
z_slices = max(ct_volumes[0].shape[2], pet_volumes[0].shape[2]) - 1  # 최대 Z 슬라이스 수

@interact(z_slice=widgets.IntSlider(min=0, max=z_slices, step=1, value=0))
def visualize_ct_and_pet_slices(z_slice):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    visualize_slices(ct_volumes[0], z_slice)
    plt.title('CT Slice')

    plt.subplot(1, 2, 2)
    visualize_slices(pet_volumes[0], z_slice)
    plt.title('PET Slice')

    plt.show()
```

```python
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt

def load_dicom_volume(folder_path):
    slices = [pydicom.dcmread(os.path.join(folder_path, file)) for file in os.listdir(folder_path) if file.lower().endswith('.dcm')]
    slices.sort(key=lambda x: float(x.ImagePositionPatient[2]))

    volume = np.stack([s.pixel_array for s in slices], axis=-1)

    return volume, slices[0].PixelSpacing[0], slices[0].PixelSpacing[1]

def compare_resolution(ct_directory, pet_directory):
    ct_resolutions = {}
    pet_resolutions = {}

    for subfolder in os.listdir(ct_directory):
        subfolder_path_ct = os.path.join(ct_directory, subfolder)
        subfolder_path_pet = os.path.join(pet_directory, subfolder)

        if os.path.isdir(subfolder_path_ct) and os.path.isdir(subfolder_path_pet):
            ct_volume, ct_pixel_spacing_x, ct_pixel_spacing_y = load_dicom_volume(subfolder_path_ct)
            pet_volume, pet_pixel_spacing_x, pet_pixel_spacing_y = load_dicom_volume(subfolder_path_pet)

            ct_resolutions[subfolder] = (ct_pixel_spacing_x, ct_pixel_spacing_y)
            pet_resolutions[subfolder] = (pet_pixel_spacing_x, pet_pixel_spacing_y)

            if (ct_pixel_spacing_x, ct_pixel_spacing_y) != (pet_pixel_spacing_x, pet_pixel_spacing_y):
                print(f"Resolution mismatch in {subfolder}: CT - {ct_pixel_spacing_x} mm x {ct_pixel_spacing_y} mm, PET - {pet_pixel_spacing_x} mm x {pet_pixel_spacing_y} mm")

    return ct_resolutions, pet_resolutions

# 대상 디렉터리 지정

ct_directory = os.path.join(r'F:\up_under_arm\upper\ct', patient_id)
pet_directory = os.path.join(r'F:\up_under_arm\upper\pet', patient_id)

# CT 및 PET 해상도 비교
ct_resolutions, pet_resolutions = compare_resolution(ct_directory, pet_directory)
```

```python
import os
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk

def load_dicom_volume(folder_path):
    slices = [pydicom.dcmread(os.path.join(folder_path, file)) for file in os.listdir(folder_path) if file.lower().endswith('.dcm')]
    slices.sort(key=lambda x: float(x.ImagePositionPatient[2]))

    volume = np.stack([s.pixel_array for s in slices], axis=-1)
    pixel_spacing = (slices[0].PixelSpacing[0], slices[0].PixelSpacing[1])

    return volume, pixel_spacing

def resample_volume(volume, target_pixel_spacing):
    current_pixel_spacing = (1.0, 1.0)  # 현재 CT 및 PET의 기본 해상도

    # 각 축에 대한 resampling 비율 계산
    resampling_ratio = (current_pixel_spacing[0] / target_pixel_spacing[0], current_pixel_spacing[1] / target_pixel_spacing[1], 1.0)

    # 보간(interpolation)을 사용하여 해상도를 조절
    resampled_volume = sitk.GetImageFromArray(volume)
    resampled_volume.SetSpacing((target_pixel_spacing[0], target_pixel_spacing[1], 1.0))
    resampled_volume = sitk.Resample(resampled_volume, (int(volume.shape[0] * resampling_ratio[0]), int(volume.shape[1] * resampling_ratio[1]), volume.shape[2]))

    return sitk.GetArrayFromImage(resampled_volume)

def overlay_pet_on_ct(ct_volume, pet_volume):
    # SimpleITK 형식으로 변환
    ct_image = sitk.GetImageFromArray(ct_volume)
    pet_image = sitk.GetImageFromArray(pet_volume)

    # CT와 PET를 같은 공간으로 정렬
    ct_image.SetSpacing((1.0, 1.0, 1.0))
    pet_image.SetSpacing((1.0, 1.0, 1.0))
    transform = sitk.CenteredTransformInitializer(ct_image, pet_image, sitk.Euler3DTransform())
    pet_image = sitk.Resample(pet_image, ct_image, transform)

    # PET를 CT 위에 overlay
    overlay_image = sitk.LabelOverlay(ct_image, pet_image)

    # Overlay된 이미지를 numpy 배열로 변환
    overlay_array = sitk.GetArrayFromImage(overlay_image)

    return overlay_array

def overlay_pet_on_ct_for_folder(ct_folder, pet_folder, target_pixel_spacing):
    ct_volumes = {}
    pet_volumes = {}
    overlay_volumes = {}

    for subfolder in os.listdir(ct_folder):
        subfolder_path_ct = os.path.join(ct_folder, subfolder)
        subfolder_path_pet = os.path.join(pet_folder, subfolder)

        if os.path.isdir(subfolder_path_ct) and os.path.isdir(subfolder_path_pet):
            # CT 및 PET 볼륨 로드
            ct_volume, _ = load_dicom_volume(subfolder_path_ct)
            pet_volume, _ = load_dicom_volume(subfolder_path_pet)

            # CT와 PET의 해상도를 대상 해상도로 조절
            resampled_ct_volume = resample_volume(ct_volume, target_pixel_spacing)
            resampled_pet_volume = resample_volume(pet_volume, target_pixel_spacing)

            # PET를 CT에 overlay
            overlay_image = overlay_pet_on_ct(resampled_ct_volume, resampled_pet_volume)

            # 결과 저장
            ct_volumes[subfolder] = resampled_ct_volume
            pet_volumes[subfolder] = resampled_pet_volume
            overlay_volumes[subfolder] = overlay_image

    return ct_volumes, pet_volumes, overlay_volumes

# 대상 디렉터리 지정
patient_id = '0090122'
ct_directory = os.path.join(r'F:\up_under_arm\upper\ct', patient_id)
pet_directory = os.path.join(r'F:\up_under_arm\upper\pet', patient_id)

# 대상 해상도 설정
target_pixel_spacing = (1.0, 1.0)  # 예시로 대상 해상도를 (1.0 mm x 1.0 mm)로 설정

# CT와 PET 볼륨 로드 및 대상 해상도로 조절 및 overlay
ct_volumes, pet_volumes, overlay_volumes = overlay_pet_on_ct_for_folder(ct_directory, pet_directory, target_pixel_spacing)

# 시각화 (예시로 하나의 폴더만 시각화)
folder_to_visualize = list(ct_volumes.keys())[0]

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(ct_volumes[folder_to_visualize][:, :, ct_volumes[folder_to_visualize].shape[2] // 2], cmap='gray', origin='lower', aspect='auto')
plt.title(f'Resampled CT Slice - {folder_to_visualize}')

plt.subplot(1, 3, 2)
plt.imshow(pet_volumes[folder_to_visualize][:, :, pet_volumes[folder_to_visualize].shape[2] // 2], cmap='hot', origin='lower', aspect='auto')
plt.title(f'Resampled PET Slice - {folder_to_visualize}')

plt.subplot(1, 3, 3)
plt.imshow(overlay_volumes[folder_to_visualize][:, :, overlay_volumes[folder_to_visualize].shape[2] // 2], cmap='hot', origin='lower', aspect='auto')
plt.title(f'PET Overlay on CT - {folder_to_visualize}')

plt.show()
```
